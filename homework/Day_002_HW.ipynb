```python
#找出擁有車子的人,回傳TURE,FALSE
app_train["FLAG_OWN_CAR"] == 'Y'
#找出擁有車子的人,回傳df
app_train[app_train["FLAG_OWN_CAR"] == 'Y']

#有車子
task1 = app_train["FLAG_OWN_CAR"] == 'Y'
#為女性
task2 = app_train["CODE_GENDER"] == 'M'
#有車子且為女性
app_train[(task1 & task2)] 
#有車子或為女性
app_train[(task1 | task2)] 
```
